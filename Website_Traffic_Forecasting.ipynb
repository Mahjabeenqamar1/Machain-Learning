## Website Traffic Forecasting using Python
thecleverprogrammer.com.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm

data = pd.read_csv("/content/drive/MyDrive/Data_set/Thecleverprogrammer.csv")
print(data.head())
```

```python
data["Date"] = pd.to_datetime(data["Date"], 
                              format="%d/%m/%Y")
print(data.info())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.plot(data["Date"], data["Views"])
plt.title("Daily Traffic of Thecleverprogrammer.com")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data["Views"], model='multiplicative', period=30)
fig = plt.figure()
# result = seasonal_decompose(data["Views"], 
#                             model='multiplicative', 
#                             freq = 30)
# fig = plt.figure()  
fig = result.plot()  
fig.set_size_inches(15, 10)
```

```python
pd.plotting.autocorrelation_plot(data["Views"])
```

```python
p, d, q = 5, 1, 2
model=sm.tsa.statespace.SARIMAX(data['Views'],
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

```python
predictions = model.predict(len(data), len(data)+50)
print(predictions)
```

```python
data["Views"].plot(legend=True, label="Training Data", 
                   figsize=(15, 10))
predictions.plot(legend=True, label="Predictions")
```
